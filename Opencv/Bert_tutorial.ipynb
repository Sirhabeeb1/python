```python
import tensorflow_hub as hub
import pandas as pd
import tensorflow_text as text
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encoder_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
text = ["I love food", "i love using python to code"]
text_preprocess = bert_preprocess_model(text)
text_preprocess.keys()
```

```python
text_preprocess['input_mask']
```

```python
text_preprocess['input_type_ids']
```

```python
text_preprocess['input_word_ids']
```

BERT uses CLS as a special token at the begining of each sentence, whereas SEP as a special token to end sentence

```python
bert_model = hub.KerasLayer(encoder_url)
```

```python
bert_result = bert_model(text_preprocess)
bert_result.keys()
```

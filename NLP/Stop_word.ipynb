```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("We just opened our wings, the flying part is coming soon")

for token in doc:
    if not token.is_stop:
        print(token) 
```

```python
def preprocess(test):
    doc = nlp(test )
    dooc_token = [ token.text for token in doc if not token.is_stop]
    return  dooc_token
            
```

```python
preprocess("i hate to much problem with excessive stress and fatique")
```

```python
df["contents_new"] = df["contents"].apply(preprocess)
```

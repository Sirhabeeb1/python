```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
sns.set()
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\py-master\ML\1_linear_reg\homeprices.csv")
df
```

```python
#plotting scatter plot for graphical representation of area and price
plt.scatter(df["area"], df["price"], color="red", marker="+" )
plt.xlabel("Area(sqr ft)")
plt.ylabel("Price($)")
plt.show()
```

```python
#creating and training model
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df['price'])
```

formular for linear regression
y = mx + b represent
m = gradient or coefficent
x = independent
y = dependent
b = intercept

```python
#the coefficient 
reg.coef_
```

```python
#the intercept
reg.intercept_
```

```python
plt.scatter(df["area"], df["price"], color="red", marker="+" )
plt.plot(df.area, reg.predict(df[["area"]]))
plt.xlabel("Area(sqr ft)")
plt.ylabel("Price($)")
plt.show()
```

```python
#Predicting price of area imported pandas 
area = pd.read_csv(r"C:\Users\HP\Desktop\py-master\ML\1_linear_reg\areas.csv")
print(area.head())
```

```python
#predicting the price using train model
p =reg.predict(area)
p
```

```python
#creating a table for predicted price
area["Price"] = p
area
```

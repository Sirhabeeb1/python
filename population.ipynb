```python
#selecting the needed columns
df1 = df[['Rank','Continent','1970 Population','1980 Population', '1990 Population', 
          '2000 Population', '2010 Population','2015 Population', '2020 Population','2022 Population',
        'Area (km²)', 'Density (per km²)', 'Growth Rate',
       'World Population Percentage']]
```

```python
#grouping using 2022 Population
df2 = df1.groupby('Continent').mean().sort_values(by= '2022 Population', ascending= False)
```

```python
#transposing it 
df2 = df2.transpose()
df2
```

```python
#extracting the needed columns
df3 = df[['Continent','1970 Population','1980 Population', '1990 Population',
           '2000 Population', '2010 Population','2015 Population', '2020 Population',
           '2022 Population']]
```

```python
#group by population
df3 = df3.groupby("Continent").mean().sort_values(by="2022 Population", ascending=False)
df3
```

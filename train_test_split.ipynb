```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
#importing the data
df = pd.read_csv(r"C:\Users\HP\Desktop\py-master\ML\6_train_test_split\carprices.csv")
```

```python
df
```

```python
#plotting graph mileage against price
plt.figure(figsize=[5,3])
plt.scatter(df["Mileage"], df["Sell Price($)"])
plt.xlabel("Mileage")
plt.ylabel("Sell Price($)")
```

```python
#plotting graph Age against price
plt.figure(figsize=[5,3])
plt.scatter(df["Age(yrs)"], df["Sell Price($)"])
plt.xlabel("Age(yrs)")
plt.ylabel("Sell Price($)")
```

```python
#creating target and feature variable
x = df.drop(columns= "Sell Price($)")
y = df["Sell Price($)"]
```

```python
#Train and split using sklearn
X_train, x_test, Y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
reg = LinearRegression()
reg.fit(X_train, Y_train)
reg
```

```python
#checking the accuracy
reg.score(X_train,Y_train)
```

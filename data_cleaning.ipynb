```python
import pandas as pd
```

```python
#importing file from my local folder
df = pd.read_excel(r'C:\Users\HP\Desktop\Customer Call List.xlsx')
df
```

```python
#getting infor of my data
df.info()
```

```python
#first dropping diplicated data
df = df.drop_duplicates()
df
```

```python
#stripping the signs before the number
df["Last_Name"] = df["Last_Name"].astype(str).str.strip("123._/")
df
```

```python
#droping a column 
df.drop(columns= "Not_Useful_Column")
```

```python
#editing the phone number to same pattern using regular expression
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
#converting the object to string and formatting it to phone format
df["Phone_Number"] =df["Phone_Number"].astype(str)
df["Phone_Number"] = df["Phone_Number"].apply(lambda num: f"{num[:3]}-{num[3:6]}-{num[6:]}")
df
```

```python
#remove non phone number 
df["Phone_Number"] =df["Phone_Number"].str.replace("nan--", "")
df["Phone_Number"] =df["Phone_Number"].str.replace("Na--", "")
df
```

```python
df[["Street_Address", "State", "Zip_code"]] = df["Address"].str.split(",", expand=True, n =2)
df
```

```python
#replace YEs with Y and No with N
df["Paying Customer"] =df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] =df["Paying Customer"].str.replace("No","N")
df
```

```python
#replacing Yes with Y and No and NO
df["Do_Not_Contact"] =df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] =df["Do_Not_Contact"].str.replace("No","N")
df
```

```python
df = df.fillna("")
df
```

```python
#dropping contact with Y(yes)
df = df.drop(df[df["Do_Not_Contact"] == "Y"].index)
```

```python
df
```

```python
#droping column with no number
df = df.drop(df[df["Phone_Number"]==""].index)
```

```python
df
```

```python
# resetting the index
df = df.reset_index(drop=True)
```

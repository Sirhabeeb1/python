```python
#droping the missing data in total bedroom and checking the shape as compared to initial one 
data_1 = data.dropna(subset=["total_bedrooms"])
data_1.shape
```

```python
# dealing with categorical data and converting them to numerical data using get dummies
dummies =pd.get_dummies(data_1.ocean_proximity, dtype="int")
dummies.head()

#cancatenate the dummies and the data 
dummies_data =pd.concat([data_1, dummies], axis="columns")
dummies_data.head()
data_clean = dummies_data.drop(["ocean_proximity", "ISLAND"], axis="columns")
data_clean.head()
```

```python
data_clean.columns
```

```python
# creting imdependent variable fro the data and removing the feature from it
x = data_clean.drop(columns=["median_house_value"])
```

```python
x.columns
```

```python
# creating the feature variable  
y = data_clean["median_house_value"]
```

```python
#importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size= 0.2,
                                                    random_state=1984)
```

```python
#importing linear regression from sklearn
from sklearn.linear_model import LinearRegression
OLS = LinearRegression()
OLS.fit(X_train, y_train)
```

```python
#printing the ols intercept
print("the OLS intercept:", str(OLS.intercept_))
```

```python
#print the ols coefficient 
print("the OLS coefficient:", str(OLS.coef_))
```

```python
# predictiing the feature(dependent variable) from X_test
y_predict = OLS.predict(X_test)
# creating a table to compared the predicted feature from actual feature 
prediction = pd.DataFrame({"PREDICTION": y_predict,"ACTUAL": y_test})
```

```python
prediction.head()
```

```python
#showing the differnec between predicted and actual value 
prediction["ERROR"] = prediction["PREDICTION"] - prediction["ACTUAL"]
print(prediction)
```

```python
#resetting the index ans having index as a column
prediction.reset_index(drop= True, inplace= True)
prediction.reset_index(inplace=True)
prediction.head()
```

```python
prediction.columns
```

```python
fig = plt.figure(figsize=(10,5))
plt.bar('index', 'ERROR', data=prediction[:70], color='black', width=0.3)
plt.ylabel('residuals')
plt.xlabel('observations')
plt.show()

```
